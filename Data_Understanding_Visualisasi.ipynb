```python
import matplotlib.pyplot as plt
import numpy as np


negara =('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 975000, 273500000, 12580000, 51780000)
kor_x = np.arange(len(negara))

plt.bar(kor_x, populasi, tick_label=negara)
plt.xticks(rotation = 90)
plt.ylabel("Populasi(Juta)")
plt.title('Populasi 10 Negara Di Dunia')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara =('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 975000, 273500000, 12580000, 51780000)
kor_x = np.arange(len(negara))

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by='Population', inplace=True)

plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation = 90)
plt.ylabel("Populasi(Juta)")
plt.title('Populasi 10 Negara Di Dunia')
plt.show()
```

```python
import matplotlib.pyplot as plt

temperatur_c = [18,21,22,20,25,28,40,32,33,45,36,38]
jam = [0,2,4,6,8,10,12,16,18,20,22,24]

plt.plot(jam, temperatur_c)
plt.ylabel("Temperatur Celcius")
plt.xlabel("Jam")
plt.title('Temperatur di Kota Salatiga')
plt.show()
```

```python
import matplotlib.pyplot as plt

temperatur_c = [18,21,22,20,25,28,40,32,33,45,36,38]
jam = [0,2,4,6,8,10,12,16,18,20,22,24]

plt.plot(jam, temperatur_c, marker='o')
plt.ylabel("Temperatur Celcius")
plt.xlabel("Jam")
plt.title('Temperatur di Kota Salatiga')
plt.show()
```

```python
import matplotlib.pyplot as plt

temperatur_c = [18,21,22,20,25,28,40,32,33,45,36,38]
temp_c_prediksi = [20,27,24,18,23,30,38,30,35,40,33,37]
jam = [0,2,4,6,8,10,12,16,18,20,22,24]

plt.plot(jam, temperatur_c, marker='o')
plt.plot(jam, temp_c_prediksi, linestyle='--')
plt.ylabel("Temperatur Celcius")
plt.xlabel("Jam")
plt.title('Temperatur di Kota Salatiga')

plt.show()
```
